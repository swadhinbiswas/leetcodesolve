```python
import pandas as pd

try:
    df = pd.read_csv('advertising.csv')
    display(df.head())
except FileNotFoundError:
    print("The file 'advertising.csv' was not found in the current directory.")
```

## Data exploration



```python
# Check dimensions
print("Shape:", df.shape)

# Concise summary
print("\nInfo:")
df.info()

# Descriptive statistics
print("\nDescribe:")
print(df.describe())

# First few rows
print("\nHead:")
display(df.head())

# Last few rows
print("\nTail:")
display(df.tail())

# Missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Potential outliers (based on IQR)
print("\nPotential Outliers (based on IQR):")
for col in ['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"\nColumn: {col}")
    print(f"Number of outliers: {len(outliers)}")
    if not outliers.empty:
        display(outliers[[col]])
```

## Data cleaning



```python
import numpy as np

df['Area Income'] = np.clip(df['Area Income'], a_min=None, a_max=df['Area Income'].quantile(0.95))

# Data type conversion
df['Timestamp'] = pd.to_datetime(df['Timestamp'])

# Duplicate removal
num_duplicates = df.duplicated().sum()
df = df.drop_duplicates()
print(f"Number of duplicate rows removed: {num_duplicates}")

# Missing value handling (recheck)
print("\nMissing Values (Recheck):")
print(df.isnull().sum())



# if df.isnull().sum().any():
#     for col in df.columns:
#         if df[col].isnull().any():
#             if pd.api.types.is_numeric_dtype(df[col]):
#                 df[col].fillna(df[col].median(), inplace=True) # Impute with median for numerical columns
#             else:
#                 df[col].fillna(df[col].mode()[0], inplace=True) # Impute with mode for categorical columns

display(df.head())
```

## Data analysis




```python
# Descriptive statistics for numerical columns
numerical_cols = ['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage']
print("Descriptive Statistics:")
display(df[numerical_cols].describe())


print("\nDistribution Analysis:")
for col in numerical_cols:
    print(f"\nColumn: {col}")
    print(df[col].describe())



print("\nRelationship with 'Clicked on Ad':")
clicked_ad_means = df.groupby('Clicked on Ad')[numerical_cols].mean()
display(clicked_ad_means)

# Compute and display the correlation matrix
print("\nCorrelation Matrix:")
correlation_matrix = df[numerical_cols + ['Clicked on Ad']].corr()
display(correlation_matrix)

# Identify strong positive/negative correlations (example threshold: 0.5)
print("\nStrong Correlations (abs > 0.5):")
strong_correlations = correlation_matrix[abs(correlation_matrix) > 0.5]
display(strong_correlations)
```

## Data visualization



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the numerical features for the histograms
numerical_features = ['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage']

# Create a figure and axes for the subplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))

# Iterate over the numerical features and create a histogram for each one
for i, feature in enumerate(numerical_features):
    row = i // 2
    col = i % 2
    ax = axes[row, col]
    sns.histplot(data=df, x=feature, ax=ax, kde=True)
    ax.set_title(f'Distribution of {feature}')
    ax.set_xlabel(feature)
    ax.set_ylabel('Frequency')

    # Add a vertical line to represent the mean
    mean_value = df[feature].mean()
    ax.axvline(mean_value, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {mean_value:.2f}')
    ax.legend()

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
